### Q: How often and how many hurricanes impact the gulf coast of the United States?

```python
#group all data by year and only use the year of the date as the index
hurricanes_year = hurricanes_df.resample('A').count().name
hurricanes_year.index = hurricanes_year.index.year
hurricanes_year.plot(marker = '*', yticks = range(0,12), xticks = range(1960,2018,2), figsize=(12,6))
plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes per Year')
plt.xticks(rotation=45)
plt.show()

#count frequency of hurricane occurences per year
hurricanes_year.value_counts().sort_index().plot(kind = 'bar',figsize=(12,6), width = 0.7, yticks=range(0,22,2))
plt.xlabel('Hurricanes per Year')
plt.ylabel('Years of Occurences')
plt.title('Frequency of Hurricane Occurences per Year')
plt.show()

#plot a cummulative distribution function of the frequency occurences per year.
hurricanes_year.hist(cumulative=True, normed=True, histtype='step', figsize=(12,6))
plt.plot([3,3],[0,1],color='red')
plt.ylabel('CDF')
plt.xlabel('Hurricanes per Year')
plt.xticks(range(0,12))
plt.yticks(np.arange(0,1.1,0.1))
plt.annotate('90%', xy=(3,0.9),xytext=(1,1), 
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

### A: 90% of the years had 3 hurricanes or less with most years having 0 or 1 hurricane per year.  This trend is pretty steady throughout history with a few outlying years, most notably 2005 with 11 hurricane impacts along the gulf coast.  These will be evaluated in greater detail later on below.

### Q: What trend or impact does coast length have on number of hurricane impacts per year?

```python
#look at hurricane impacts of state vs. state coast length
states_df = pd.DataFrame(states)
coasts_df = pd.DataFrame.from_dict(coasts_dict, orient='index')
coasts_df.columns = ['coast_length']
states_coasts = states_df.merge(coasts_df, left_index=True, right_index=True)
states_coasts.plot(kind='scatter', x='coast_length',y='hurricanes', figsize=(12,6),
                   grid=True, yticks=range(0,36,2),xticks=range(0,9000,500))
plt.xlabel('Coast Length (statute miles)')
plt.ylabel('Hurricanes')

#annotate plots with state label
for index,row in states_coasts.iterrows():
    plt.annotate(index, xy=(row['coast_length'],row['hurricanes']))


#linear fit to the scatter plot
a, b = np.polyfit(states_coasts['coast_length'], states_coasts['hurricanes'],1)
x = np.arange(0,9000,500)
plt.plot(x,x*a+b, color ='red')

plt.show()

#print the pearson correlation value
print("The correlation between coast length and number of hurricanes: {}".format(
    np.corrcoef(states_coasts['coast_length'],states_coasts['hurricanes'])[0][1]))
```

### A: After looking into hurricane counts vs. coast length for each state it seems that their is a positive correlation between the two quantities.  This suggest that states with larger coast lines tend to be hit more often than states with smaller coast lines.  This may just be circumstantial to other factors such as sea temperature, sea floor depth, etc. along those coast lines.

### Q: How do the ocean properties and hurricane impacts change over time by location?  Broken down by state?

```python
#set up dictionary for state locations stored as lat min, lat max, lon min, lon max
state_locs = {'Louisiana': np.array([28,30,-94,-89]), 'Alabama': np.array([29,30,-88,-87]),
              'Mississippi':np.array([29,30,-89,-88]), 'Texas': np.array([26,30,-97,-94]), 'Florida': np.array([25,30,-87,-80])}

#dissassociate location into lat and lon
wod_loc_split = wod_df_loc
for index, row in wod_loc_split.iterrows():
    lat, lon = index[0].split(',')
    lat = float(lat.replace('(','').strip())
    lon = float(lon.replace(')','').strip())
    
    #add lat an lon as columns to each row in wod_df_loc
    wod_loc_split.loc[(index[0],index[1]),'lat']=lat
    wod_loc_split.loc[(index[0],index[1]),'lon']=lon
    
#now drop the location from the multi index
wod_df_loc_drop = wod_loc_split
wod_df_loc_drop.index = wod_df_loc_drop.index.droplevel()
wod_df_loc_drop.sort_index(inplace=True)

#create state dfs from the wod_df_loc dataframe
states = {}
for key, arr in state_locs.items():
    states[key] = wod_df_loc_drop[(wod_df_loc_drop.lat.values >= arr[0]) & 
                                (wod_df_loc_drop.lat.values <= arr[1]) & 
                                (wod_df_loc_drop.lon.values >= arr[2]) & 
                                (wod_df_loc_drop.lon.values <= arr[3])]   


```

```python
#set the longitude and latitude boundaries of the map
extent = [-100,-80,22,34]

#set the random number generator seed
np.random.seed(42)
'''Makes the heat maps for the wod data for each year from 1960 to 2013.'''
heat_maps={}
years={}
    
#make the heatmaps per year
for year in range(1960,2013):
    #create blank dataframe to keep the wod data for each year
    year_df = pd.DataFrame(columns=states['Louisiana'].columns.values)
        
    #concat each state dataframes year info into one big year dataframe
    for df in states.values():
        year_df = pd.concat([year_df,df[df.index.year==year]])
            
    #reset the index and drop the date information and sort the dataframe by the lat and lon
    year_df = year_df.reset_index(drop=True).sort_values(['lat','lon'])
        
    #store the dataframe in a dictionary of dataframes by year
    years[year] = year_df

    #now go through the year dataframes and only use the non blank ones
    if not year_df.empty:
        #create the latitudes and longitudes for the heatmap
        lat = np.arange(extent[2],extent[3]+1)
        lon = np.arange(extent[0],extent[1]+1)
            
        #group the dataframe by lat and lon for easier recall of data
        year_df = year_df.groupby(['lat','lon']).mean()
            
        #create dictionary of heatmaps by parameter
        params={}
        
        for param in year_df.columns.values:
            #create 2D array for the heatmap data for each lat lon pair
            latlon=[[0] * len(lon) for i in range(len(lat))]
            
            
            #loop over latitudes as array rows
            lat_count = 0
            for i in lat:
                    
                #loop over longitudes as array columns
                lon_count = 0
                for j in lon:
                    '''some lat and lon pairs don't have data.  in this case 
                    pick a random number for that parameter for that year'''
                    try:
                        latlon[lat_count][lon_count] = year_df.loc[(i,j)][param]
                    except:
                        latlon[lat_count][lon_count] = np.random.choice(year_df[param])
                    lon_count += 1
                        
                lat_count += 1
                
            #store the param heat map in the params dictionary with the param as the key
            params[param] = latlon
            
        #store the params dictionary of heatmaps for all the parameters in a dictionary of heatmaps stored by year
        heat_maps[year] = params
```

```python
from IPython.core.debugger import set_trace
from IPython.html.widgets import *
import cartopy.crs as ccrs
import cartopy.feature as cfeature


#set the longitude and latitude boundaries of the map
extent = [-100,-80,22,34]

#create the categories by state dataframe
cats_by_state = pd.pivot_table(hurricanes_df, index=[hurricanes_df.index.year,'state'], 
                               columns='category', values='name', aggfunc='count', fill_value=0)
            
#calculate the max and min values for each parameter so that the colormap bars are uniform for each year and param combo
max_vals={}
min_vals={}
for param in years[1960].columns.values[:-2]:
    Max =[]
    Min =[]
    for year, df in years.items():
        Min.append(df[param].min())
        Max.append(df[param].max())
    max_vals[param]=max(Max)
    min_vals[param]=min(Min)

def make_plot(year, parameter):
    '''
    generates the cartopy map and the heatmap of the wod data given
    projection_crs is the projection style of the map to be plotted(a cartopy.crs object)
    extent contains the boundaries of the map to plotted as a list [lon_min,lon_max,lat_min,lat_max]
    heat_data is the heat map for the year and parameter of interest
    '''
    #setting map features
    projection_crs = ccrs.Miller()
    #parameter = 'temperature'
    heat_data = heat_maps.get(year,1960)[parameter]
    
    #bring up axes and set map boundaries
    fig = plt.figure()
    rect = 0.1, 0.1, 1.2, 1.2
    ax = fig.add_axes(rect, projection=projection_crs)
    ax.set_extent(extent, crs=projection_crs)
    
    

    # Create a feature for States/Admin 1 regions at 1:50m from Natural Earth
    states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        facecolor='none')
    
    #add features to the map
    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(states_provinces, edgecolor='gray')

    #set the lat and lon for the plot, create a meshgrid, and then make the heatmap from lat,lon and the heat_data
    lat = np.arange(extent[2],extent[3]+1)
    lon = np.arange(extent[0],extent[1]+1)
    Lat,Lon = np.meshgrid(lat,lon)
    
    #add the heat map to the plot
    heat = ax.pcolormesh(Lon, Lat, np.transpose(heat_data), cmap='coolwarm', 
                         vmin=min_vals[parameter], vmax=max_vals[parameter])
    fig.colorbar(heat)
    
    plt.title(parameter + ' ' + str(year) + ' heat map')
    plt.show()
    
    

    cats_by_state.loc[year].plot(kind='bar', width=1, yticks=range(0,11))
    plt.legend(loc='upper right', bbox_to_anchor=(1.25,1), fancybox=True)
    plt.ylabel('Number of Hurricanes per State')
    plt.title('Number of Hurricanes in '+str(year)+' with Category Info')
    plt.show()

#interact to make plots for years and parameters searchable
interact(make_plot, year=(1960, 2013, 1), parameter=list(states['Louisiana'].columns.values[:-2]))

    
   
```
